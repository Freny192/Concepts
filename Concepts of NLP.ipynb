```python
#1.Calculate the word frequency counts (eliminating stop words)
```

```python
tokenize_wd = nltk.word_tokenize(mytext)
print("count of words :- ",len(tokenize_wd))
tokenize_wd[:10]
```

```python
#Removing extra special characters like , space etc.
tokenizer = nltk.RegexpTokenizer(r"\w+")
remove_s_word = tokenizer.tokenize(mytext)
remove_s_word[:10]
```

```python
print("count of word :- ",len(remove_s_word))
```

```python
#converting text into the lower character for NLP procedure:
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
Token_lower = tokenizer.tokenize(mytext.lower())
print(Token_lower)
```

```python
remove_word_lower = list(map(lambda word: word.lower(),remove_s_word))
```

```python
from nltk.corpus import stopwords
```

```python
#Using Filter for tokenzing word by removing stopwords i.e. one of the important part of the NLP process.
filtered_words = [word for word in remove_word_lower if word not in stopwords.words('english')]
#filtered_words[:10]
print(filtered_words)
```

```python
print("count of word :- ",len(filtered_words))
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
#lemmatizer is a process of grouping same meaning words together.
lemmatizer = WordNetLemmatizer()
rsp_word_lower_lem = list(map(lambda word:lemmatizer.lemmatize(word),filtered_words))
rsp_word_lower_lem[:10]
```

```python
frequency_dist_words = FreqDist(rsp_word_lower_lem)
frequency_dist_words.most_common(10)
```

```python
#Plot top 10-word frequency (eliminating stop words)  
```

```python
import matplotlib.pyplot as plt
```

```python
freq_dist = nltk.FreqDist(filtered_words)
freq_dist

print(freq_dist)
print(freq_dist.most_common(10))
freq_dist.plot(10)
```

```python
plt.figure(figsize=(10,5))
barlist=plt.bar(*zip(*frequency_dist_words.most_common(10)))
barlist[0].set_color('r')
barlist[1].set_color('y')
barlist[2].set_color('g')
barlist[3].set_color('c')
plt.title("Top -10 frequent words")
plt.xlabel(" frequent Words")
plt.ylabel("Count")
plt.show()
```
