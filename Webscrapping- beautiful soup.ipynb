```python
# read the data from the URL
url= requests.get("https://www.scrapethissite.com/pages/simple/")
```

```python
# parse the URL using Beautiful Soup library 
soup=BeautifulSoup(url.content, 'html.parser') 
```

```python
country_tag = soup.find_all('div',{'class':'country-info'})
```

```python
for i, record in enumerate(country_tag):
    #print(i)
    print(record.find('span',{'class':'country-capital'}).get_text())
    print(record.find('span',{'class':'country-population'}).get_text())
    print(record.find('span',{'class':'country-area'}).get_text())
    #print(record.h3.a['title'])
    print('')
```
